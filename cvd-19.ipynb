После острейшей волны начала 2022 наблюдается устойчивый нисходящий тренд: к середине–концу 2023 уровень новых случаев упал к десяткам тысяч, и в 2024 держится даже ниже 50 000 в пике.

```python

latest = df.sort_values('date').groupby('location').last()

top_cases = latest.sort_values('total_cases', ascending=False).head(10)
top_deaths = latest.sort_values('total_deaths', ascending=False).head(10)
top_vacc = latest.sort_values('total_vaccinations', ascending=False).head(10)


def plot_bar(data, column, title):
    plt.figure(figsize=(10, 6))
    sns.barplot(x=data[column], y=data.index, palette="viridis")
    plt.title(title)
    plt.xlabel(column.replace('_', ' ').title())
    plt.tight_layout()
    plt.show()

plot_bar(top_cases, 'total_cases', 'Top-10 стран по числу заражений')
plot_bar(top_deaths, 'total_deaths', 'Top-10 стран по числу смертей')
plot_bar(top_vacc, 'total_vaccinations', 'Top-10 стран по числу вакцинаций')
```

```python
def plot_country_trend(country, column):
    country_df = df[df['location'] == country]
    plt.figure(figsize=(12, 5))
    plt.plot(country_df['date'], country_df[column])
    plt.title(f"{column.replace('_', ' ').title()} в {country}")
    plt.xlabel("Дата")
    plt.ylabel(column.replace('_', ' ').title())
    plt.grid(True)
    plt.tight_layout()
    plt.show()


countries = ['Russia', 'United States', 'Germany', 'India', 'Brazil']
columns = ['new_cases', 'new_deaths', 'new_vaccinations']

for country in countries:
    for col in columns:
        plot_country_trend(country, col)
```

```python
from prophet import Prophet
```

```python
russia_df = df[df['location'] == 'Russia'][['date', 'new_cases']].copy()
```

```python
russia_df = russia_df.dropna()
russia_df = russia_df[russia_df['new_cases'] >= 0]
```

```python
russia_df.rename(columns={'date': 'ds', 'new_cases': 'y'}, inplace=True)
```

```python
model = Prophet()
model.fit(russia_df)
```

```python
future = model.make_future_dataframe(periods=14)
forecast = model.predict(future)
```

```python
fig1 = model.plot(forecast)
plt.title("Прогноз новых случаев COVID-19 в России")
plt.xlabel("Дата")
plt.ylabel("Новые случаи")
plt.tight_layout()
plt.show()

```

```python
fig2 = model.plot_components(forecast)
plt.tight_layout()
plt.show()
```

```python

from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

```

```python
russia_df = df[df['location'] == 'Russia'][['date', 'new_cases']].copy()
russia_df = russia_df.dropna()
russia_df = russia_df[russia_df['new_cases'] >= 0]
russia_df.rename(columns={'date': 'ds', 'new_cases': 'y'}, inplace=True)
```

```python
split_date = '2022-12-01'
train = russia_df[russia_df['ds'] < split_date]
test = russia_df[russia_df['ds'] >= split_date]
```

```python
model = Prophet()
model.fit(train)
```

```python
future = model.make_future_dataframe(periods=len(test), freq='D')
forecast = model.predict(future)
```

```python
predicted = forecast[['ds', 'yhat']].set_index('ds').loc[test['ds']]
actual = test.set_index('ds')
```

```python
mae = mean_absolute_error(actual['y'], predicted['yhat'])
mape = mean_absolute_percentage_error(actual['y'], predicted['yhat']) * 100
print(f" MAE:  {mae:.0f}")
print(f" MAPE: {mape:.2f}%")
```

```python
combined = actual.join(predicted)
combined = combined[combined['y'] != 0]
```

```python
mae = mean_absolute_error(combined['y'], combined['yhat'])
mape = mean_absolute_percentage_error(combined['y'], combined['yhat']) * 100
print(f" MAE:  {mae:.0f}")
print(f" MAPE: {mape:.2f}%")
```

В среднем модель ошибается на 77 тыс. новых случаев в день.
Очень высокая процентная ошибка — это означает, что в дни с низкими значениями предсказания могли быть сильно завышены.
